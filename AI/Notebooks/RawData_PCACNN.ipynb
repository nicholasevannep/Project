```python
from keras.models import Sequential
from keras.layers import Dense, Input, Dropout, Flatten
from keras.layers.convolutional import Conv1D, MaxPooling1D
from keras import backend as k
from keras.utils import to_categorical
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA 
```

```python
df = pd.read_csv("rawdata.csv")
feature = df.iloc[:, 1:10002]
target = df["STATUS"]
```

```python
pca = PCA(n_components=1)
pricipalComponents = pca.fit_transform(feature)
dfPCA = pd.DataFrame(data=pricipalComponents, columns=['features'])
dfPCA
```

```python
x_train, x_test, y_train, y_test = train_test_split(dfPCA, target, train_size=0.6, random_state = 0)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
x_train = x_train.values.reshape(-1, 1, 1)
x_test = x_test.values.reshape(-1, 1, 1)
```

```python
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
def cnn_model_evaluation(x_train, y_train, x_test, y_test):
    epoch, verbose, batch_size = 10, 0, 36
    # cnnModel = Sequential()
    # cnnModel.add(Conv1D(16, kernel_size=3, activation='relu', input_shape=(10002, 1)))
    # cnnModel.add(Conv1D(16, kernel_size=3, activation='relu'))
    # cnnModel.add(MaxPooling1D(pool_size=4))
    # cnnModel.add(Dropout(0.2))
    # cnnModel.add(Conv1D(32, kernel_size=3, activation='relu'))
    # cnnModel.add(MaxPooling1D(pool_size=4))
    # cnnModel.add(Flatten())
    # cnnModel.add(Dropout(0.2))
    # cnnModel.add(Dense(300, activation='relu'))
    # cnnModel.add(Dense(3, activation='softmax'))
    cnnModel = Sequential()
    cnnModel.add(Conv1D(16, kernel_size=1, activation='relu', input_shape=(1, 1)))
    cnnModel.add(Conv1D(16, kernel_size=1, activation='relu'))
    cnnModel.add(Dropout(0.2))
    cnnModel.add(MaxPooling1D(pool_size=1))
    cnnModel.add(Flatten())
    cnnModel.add(Dense(300, activation='relu'))
    cnnModel.add(Dense(3, activation='softmax'))
    cnnModel.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    cnnModel.fit(x_train, y_train, epochs=epoch, batch_size=batch_size, verbose=verbose)
    _, acc = cnnModel.evaluate(x_test, y_test, batch_size=batch_size, verbose=0)
    # cnnModel.summary()
    return acc
```

```python
iteration = 10
for r in range(iteration):
    score = cnn_model_evaluation(x_train, y_train, x_test, y_test)
    print(f'Iteration : {r}, Accuary : {score * 100.0}%')
```
