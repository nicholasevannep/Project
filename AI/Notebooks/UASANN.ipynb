```
input = [
        [1,2,-1],
        [-1,3,-2]
        ]

weight= [
        [1,1,-1],
        [2,1,1],
        [-1,2,-3],
        [1,2,3],
        [1,1,3]
        ]

epoch = 1000
lr=0.5
weight_akhir = []
print(weight)
jarak1=[]
index1=[]
jarak2=[]
index2=[]

for berapalah in range(epoch):

  for i in range (len(input)):
    a=0
    b=0
    c=0
    d=0
    e=0
 
    for j in range (len(input[i])):
      a += ((input[i][j]- weight[0][j])**2)
      b += ((input[i][j]- weight[1][j])**2)
      c += ((input[i][j]- weight[2][j])**2)
      d += ((input[i][j]- weight[3][j])**2)
      e += ((input[i][j]- weight[4][j])**2)

    a= math.sqrt(a)
    b= math.sqrt(b)
    c= math.sqrt(c)
    d= math.sqrt(d)
    e= math.sqrt(e)
    jarak=min(a,b,c,d,e)
    apala = 1*math.exp(-berapalah/epoch)
    if(i==0):
      jarak1.append(jarak)
    if(i==1):
      jarak2.append(jarak)
    if(a==jarak):
      bedaa=0
      bedab=1
      bedac=2
      bedad=3
      bedae=4
      if(i==0):
        index1.append(0)
      if(i==1):
        index2.append(0)
    elif(b==jarak):
      bedaa=1
      bedab=0
      bedac=1
      bedad=2
      bedae=3
      if(i==0):
        index1.append(1)
      if(i==1):
        index2.append(1)
    elif(c==jarak):
      bedaa=2
      bedab=1
      bedac=0
      bedad=1
      bedae=2
      if(i==0):
        index1.append(2)
      if(i==1):
        index2.append(2)
    elif(d==jarak):
      bedaa=3
      bedab=2
      bedac=1
      bedad=0
      bedae=1
      if(i==0):
        index1.append(3)
      if(i==1):
        index2.append(3)
    elif(e==jarak):
      bedaa=4
      bedab=3
      bedac=2
      bedad=1
      bedae=0
      if(i==0):
        index1.append(4)
      if(i==1):
        index2.append(4)

    if(a==jarak or bedaa==1):
      new_weight=[]
      for x in range(len(input[i])):
        new_weight.append(((input[i][x]-weight[0][x])*lr*math.exp((-(bedaa)**2)/(2*(apala)**2)))+weight[0][x])
      weight[0]=new_weight

    if(b==jarak or bedab==1):
      new_weight=[]
      for x in range(len(input[i])):
        new_weight.append(((input[i][x]-weight[1][x])*lr*math.exp((-(bedab)**2)/(2*(apala)**2)))+weight[1][x])
      weight[1]=new_weight

    if(c==jarak or bedac==1):
      new_weight=[]
      for x in range(len(input[i])):
        new_weight.append(((input[i][x]-weight[2][x])*lr*math.exp((-(bedac)**2)/(2*(apala)**2)))+weight[2][x])
      weight[2]=new_weight

    if(d==jarak or bedad==1):
      new_weight=[]
      for x in range(len(input[i])):
        new_weight.append(((input[i][x]-weight[3][x])*lr*math.exp((-(bedad)**2)/(2*(apala)**2)))+weight[3][x])
      weight[3]=new_weight
 
    if(e==jarak or bedae==1):
      new_weight=[]
      for x in range(len(input[i])):
        new_weight.append(((input[i][x]-weight[4][x])*lr*math.exp((-(bedae)**2)/(2*(apala)**2)))+weight[4][x])
      weight[4]=new_weight
    
  if(weight==weight_akhir):
    print(weight_akhir)
    print(berapalah)  
    break
  weight_akhir=weight[:]
  print(weight)

```

```
cov_mat = np.cov([samples[0,:],samples[1,:],samples[2,:],samples[3,:]])
print('Covariance Matrix:\n', cov_mat)
```

```
# eigenvectors and eigenvalues for the from the scatter matrix
eig_val_sc, eig_vec_sc = np.linalg.eig(scatter_matrix)

# eigenvectors and eigenvalues for the from the covariance matrix
eig_val_cov, eig_vec_cov = np.linalg.eig(cov_mat)

for i in range(len(eig_val_sc)):
    eigvec_sc = eig_vec_sc[:,i].reshape(1,4).T
    eigvec_cov = eig_vec_cov[:,i].reshape(1,4).T
    assert eigvec_sc.all() == eigvec_cov.all(), 'Eigenvectors are not identical'

    print('Eigenvector {}: \n{}'.format(i+1, eigvec_sc))
    print('Eigenvalue {} from scatter matrix: {}'.format(i+1, eig_val_sc[i]))
    print('Eigenvalue {} from covariance matrix: {}'.format(i+1, eig_val_cov[i]))
    print('Scaling factor: ', eig_val_sc[i]/eig_val_cov[i])
    print(40 * '-')
```

```
import numpy as np
import idx2numpy
import matplotlib.pyplot as plt
import tensorflow as tf


x_train = idx2numpy.convert_from_file('train-images.idx3-ubyte')
y_train = idx2numpy.convert_from_file('train-labels.idx1-ubyte')
x_test = idx2numpy.convert_from_file('t10k-images.idx3-ubyte')
y_test = idx2numpy.convert_from_file('t10k-labels.idx1-ubyte')

plt.imshow(x_train[0], cmap=plt.cm.binary)
print(y_train[0])
```

refrence : https://stackoverflow.com/questions/40427435/extract-images-from-idx3-ubyte-file-or-gzip-via-python

```
print(x_train.shape)
```

```
x_train = x_train / 255.0
x_test = x_test / 255.0
x_train = x_train[..., tf.newaxis]
x_test = x_test[..., tf.newaxis]

y_train = tf.keras.utils.to_categorical(y_train, 10)
y_test = tf.keras.utils.to_categorical(y_test, 10)
```

```
x_train = np.pad(x_train,((0,0),(2,2),(2,2),(0,0)), 'constant')
x_test = np.pad(x_test,((0,0),(2,2),(2,2),(0,0)), 'constant')

print("new shape :{}".format(x_train.shape))
```

ref : https://colab.research.google.com/drive/1CVm50PGE4vhtB5I_a_yc4h5F-itKOVL9#scrollTo=1w66ueiLlP0k

```

model = tf.keras.Sequential() 

model.add(tf.keras.layers.Conv2D(6, kernel_size=(5, 5), activation='relu', input_shape=(32,32,1)))

model.add(tf.keras.layers.AveragePooling2D(pool_size=(2,2),padding='valid'))

model.add(tf.keras.layers.Conv2D(16, (5, 5), activation='relu')) 

model.add(tf.keras.layers.AveragePooling2D(pool_size=(2,2),padding='valid'))

model.add(tf.keras.layers.Conv2D(120, kernel_size=(5, 5), activation='relu'))

model.add(tf.keras.layers.Flatten())

model.add(tf.keras.layers.Dense(84, activation='relu'))

model.add(tf.keras.layers.Dense(10, activation='softmax'))

model.compile(loss=tf.keras.losses.categorical_crossentropy, optimizer=tf.keras.optimizers.Adadelta(), metrics=['accuracy'])

model.summary()
```

```
themodel=model.fit(x_train, y_train, batch_size=128, epochs=10, verbose=1, validation_data=(x_test, y_test))
```

```
score = model.evaluate(x_test, y_test, verbose=0) 
print('Test loss:', score[0]) 
print('Test accuracy:', score[1])
```

```
plt.title('loss & accuracy')
plt.plot(themodel.history['loss'], label='Train')
plt.plot(themodel.history['val_loss'], label='Test')
plt.plot(themodel.history['acc'], label='Train')
plt.plot(themodel.history['val_acc'], label='Test')
plt.show()
```
