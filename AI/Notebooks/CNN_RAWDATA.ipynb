```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
raw_dataset = pd.read_csv('rawdata.csv')
target = raw_dataset[["STATUS"]]
feature = raw_dataset.drop(['STATUS', 'SAMPLE_ID'], axis = 1)
raw_dataset
```

```python
target['STATUS'].astype('float')
```

```python
from sklearn.preprocessing import MinMaxScaler
# define min max scaler
scaler = MinMaxScaler()
# transform data
scaled = scaler.fit_transform(target)
```

```python
target=scaled
```

```python
x_train, x_test, y_train, y_test = train_test_split(feature, target, train_size=0.8, random_state = 0)
```

```python
print("x-train shape : {}".format(x_train.shape))
print("x-test shape : {}".format(x_test.shape))
print("y-train shape : {}".format(y_train.shape))
print("y-test shape : {}".format(y_test.shape))
```

```python
x_train = x_train.values.reshape(-1, 10001, 1)
x_test = x_test.values.reshape(-1, 10001, 1)
```

```python
# y_train = tf.keras.utils.to_categorical(y_train, 10)
# y_test = tf.keras.utils.to_categorical(y_test, 10)
```

```python
x_train = x_train / 255.0
x_test = x_test / 255.0
x_train = x_train[..., tf.newaxis]
x_test = x_test[..., tf.newaxis]
```

```python
# x_train = np.pad(x_train,((0,0),(2,2),(2,2),(0,0)), 'constant')
# x_test = np.pad(x_test,((0,0),(2,2),(2,2),(0,0)), 'constant')

print("new x-train shape : {}".format(x_train.shape))
print("new x-test shape : {}".format(x_test.shape))
print("new y-train shape : {}".format(y_train.shape))
print("new y-test shape : {}".format(y_test.shape))
```

```python
model = tf.keras.Sequential() 
model.add(tf.keras.layers.Conv1D(16, (1), padding='same', activation='relu', kernel_initializer='he_uniform', input_shape=(10001,1)))
model.add(tf.keras.layers.MaxPooling1D((1)))
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(10, activation='relu', kernel_initializer='he_uniform'))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
opt = tf.keras.optimizers.SGD(lr=0.01, momentum=0.9)
model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
themodel=model.fit(x_train, y_train, batch_size=36, epochs=10, verbose=0, validation_split=0.6, validation_data=(x_test, y_test))
```

```python
score = model.evaluate(x_test, y_test, verbose=0) 
print('Test Loss:', score[0]) 
print('Test Accuracy:', score[1])
```
