Reading Image and apply it ino x_train_img and x_test_img

```
def image_process(src,path):
  images=[]
  for i in src:
    img = cv2.imread(os.path.join(path,i))
    img = cv2.resize(img, (75, 115))
    img = img.astype(np.float32)/255
    images.append(img)
  return images

#src = df["image"]
#path=/content/iris_multimodal_dataset/images
```

```
x_train_img=image_process(x_train_img,"/content/iris_multimodal_dataset/images")
```

```
x_test_img=image_process(x_test_img,"/content/iris_multimodal_dataset/images")
```

```

from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.models import Model, load_model
from keras.layers import Embedding, LSTM, Dropout, Dense, Input, Bidirectional, Flatten, Conv2D, MaxPooling2D, concatenate, Conv1D, MaxPooling1D
import keras.backend as K
from keras.optimizers import Adam, RMSprop
from keras.callbacks import ModelCheckpoint, CSVLogger, EarlyStopping
```

Change Image into Array

```
x_train_img=np.array(x_train_img)
```

```
x_test_img=np.array(x_test_img)
```

Deep Learning model combination of CNN and BPNN

```
num_input = Input(shape=(x_train.shape[1],),name='main_input')
num = Dense(32)(num_input)
num_out=Dense(3,activation='sigmoid',name='num_output')(num)


cnn_input = Input(shape=(x_train_img.shape[1], x_train_img.shape[2], x_train_img.shape[3]),name='cnn_input' )
y = Conv2D(32, (3, 3), activation='relu', input_shape=(x_train_img.shape[1], x_train_img.shape[2], x_train_img.shape[3]))(cnn_input)
y = MaxPooling2D(2, 2)(y)
y = Conv2D(64, (3, 3), activation='relu')(y)
y = MaxPooling2D(2, 2)(y)
y = Conv2D(128, (3, 3), activation='relu')(y)
y = MaxPooling2D(2, 2)(y)
y = Conv2D(128, (3, 3), activation='relu')(y)
y = MaxPooling2D(2, 2)(y)
y = Flatten()(y)
y = Dropout(0.3)(y)
cnn_out = Dense(3, activation='relu')(y)

x=concatenate([num_out,cnn_out])
x = Dense(64, activation='relu')(x)
x = Dense(64, activation='relu')(x)
x = Dense(64, activation='relu')(x)

main_output = Dense(3, activation='sigmoid', name='main_output')(x)

```

```
model = Model(inputs=[num_input,cnn_input],outputs=main_output)
model.summary
```

```
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```
