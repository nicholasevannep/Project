```
from imutils import paths
import numpy as np
import argparse
import imutils
import cv2
```

```
from google.colab.patches import cv2_imshow
```

Reading Input image and append it to images array

```
images=[]
imga=cv2.imread("/content/1a.jpg")
images.append(imga)
imgb=cv2.imread("/content/1b.jpg")
images.append(imgb)
imgc=cv2.imread("/content/1c.jpg")
images.append(imgc)
```

Using cv2.createStitcher() function for stitching image


```
stitcher = cv2.createStitcher() if imutils.is_cv3() else cv2.Stitcher_create()
(status, stitched) = stitcher.stitch(images)
```

```
# use cv2.imshow(stitched) if not using google collab
cv2_imshow(stitched)
```
