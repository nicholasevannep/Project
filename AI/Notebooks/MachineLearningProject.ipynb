```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import statsmodels.formula.api as sm
import seaborn as sns


data = pd.read_csv('E202-COMP7117-TD01-00 - classification.csv')

for i in range(len(data)):
  if data['free sulfur dioxide'][i] == 'High':
    data['free sulfur dioxide'][i] = 3
  elif data['free sulfur dioxide'][i] == 'Medium':
    data['free sulfur dioxide'][i] = 2
  elif data['free sulfur dioxide'][i] == 'Low':
    data['free sulfur dioxide'][i] = 1
  else:
    data['free sulfur dioxide'][i] = 0

  if data['density'][i] == 'Very High':
    data['density'][i] = 0
  elif data['density'][i] == 'High':
    data['density'][i] = 3
  elif data['density'][i] == 'Medium':
    data['density'][i] = 2
  elif data['density'][i] == 'Low':
    data['density'][i] = 1

  if data['pH'][i] == 'Very Basic':
    data['pH'][i] = 3
  elif data['pH'][i] == 'Normal':
    data['pH'][i] = 2
  elif data['pH'][i] == 'Very Acidic':
    data['pH'][i] = 1
  else:
    data['pH'][i] = 0

  if data['quality'][i] == 'Great':
    data['quality'][i] = 4
  elif data['quality'][i] == 'Good':
    data['quality'][i] = 3
  elif data['quality'][i] == 'Fine':
    data['quality'][i] = 2
  elif data['quality'][i] == 'Decent':
    data['quality'][i] = 1
  elif data['quality'][i] == 'Fair':
    data['quality'][i] = 0
  

data
```
